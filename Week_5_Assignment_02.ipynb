# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.


Arithmatic Error is thrown when an error occurs while performing __mathematical operations.__

This errors include attempting to perform a __bitshift by a negative amount, and any call to intdiv() that would result in a value outside the possible bounds of an int.__

    -ZeroDivisionError
    -OverflowError
    -FloatingPointError

```python
# Handling ZeroDivisionError

try:
    a = 5/0
    print(a)
except ArithmeticError as e:
    print(e)
```

```python
# Handling OverflowError

j = 5.0
try:
    for i in range(1,10000):
        j = j ** i
except ArithmeticError as e:
    print(f'{e}, {e.__class__}')
```

# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.


LookupError Exception is the base class for errors raised when something can't be found.

The base class for the exceptions that are raised when a key or index used a mapping or sequence is invalid:
    
    -IndexError
    -KeyError
    
An IndexError is raised when a sequence reference is out of range

```python
lst = [1,2,3,4,5]
try:
    for i in range(10):
        print(lst[i])
except IndexError as e:
    print(e)
```

```python
dict = {'name':'Akshay',
        'age' : 21}

try:
    print(dict['city'])
except KeyError as e:
    print(e)
```
